```
# using image-quality library
img = img_as_float(io.imread('/content/airport_blur_frame313.jpg'))
b_features = brisque.calculate_features(img, 7, 7 / 6)
b_score = brisque.score(img)
print('Score: ', b_score)
print('Feature: ', b_features)
```
