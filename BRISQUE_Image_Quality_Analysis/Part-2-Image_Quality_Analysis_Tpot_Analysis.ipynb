```
'''
Image Quality Analysis/Classification using Brisque features -

This notebook uses train brisque data and identify best hyper parameter along with algorithm which gives best result to this train data. For that, we are using TPOTClassifier.

'''
```

```
!pip install tpot
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
import pandas as pd 
import numpy as np
```

```
data = pd.read_csv('/train_brisque_data.csv')
data.describe()
```

```
def replaceZeroWithMean(column):
    if column == None:
        return
    column_values_non_zero = data.loc[data[column] != 0, column]
    mean = sum(column_values_non_zero)/len(column_values_non_zero)
    data.loc[data[column] == 0 , column] = mean
```

```
columns_zero_to_mean = ['feat1', 'feat2', 'feat3','feat4','feat5','feat6', 'feat7', 'feat8','feat9', 'feat10',
                        'feat11','feat12', 'feat13', 'feat14','feat15','feat16','feat17','feat18','feat19','feat20',
                        'feat21','feat22','feat23','feat24','feat25','feat26','feat27','feat28','feat29','feat30',
                        'feat31','feat32','feat33','feat34','feat35','feat36']

for column in columns_zero_to_mean:
    replaceZeroWithMean(column)
```

```
data.describe()[columns_zero_to_mean] 
```

```
data.rename(columns = {'Outcome': 'label'}, inplace=True)
```

```
data_y = data['label']
data_X = data.drop(['label'], axis=1)
```

```
X_train, X_test, y_train, y_test = train_test_split(data_X, data_y,
                                                    train_size=0.75, test_size=0.25, random_state=0)
```

```
tpot = TPOTClassifier(verbosity=1, max_time_mins=50) 
tpot.fit(X_train, y_train)
```

```
print(tpot.score(X_test, y_test))
#tpot_scriptResult = 0.8489560195468681
#tpot_script2Result= 0.8578409595735229
```
