```
#### for blurrying 
print("creating Blurry pics for " )
image = Image.open("/content/drive/My Drive/TLFile/dog4.jpg")
corrupted_image = corrupt(image, corruption_name='defocus_blur', severity=4)
plt.imshow(corrupted_image)
plt.show()
pix = np.array(corrupted_image)
imageio.imwrite('/content/drive/dog_blur.jpg', pix) 

#For Jitter 
print("creating Jitter pics  for ")
corrupted_image1 = corrupt(image, corruption_name='impulse_noise', severity=5)
pix1 = np.array(corrupted_image1)
imageio.imwrite('/content/drive/dog_jitter.jpg' , pix1)

######For Contrast
print("creating Contrast pics  for " )
corrupted_image2 = corrupt(image, corruption_name='contrast', severity=1)
pix2 = np.array(corrupted_image2)
imageio.imwrite('/content/drive/dog_contrast.jpg', pix2)
```
